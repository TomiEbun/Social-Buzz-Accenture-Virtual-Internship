```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import calendar
from datetime import datetime
```

```python
content = pd.read_csv("Content.csv")
content
```

```python
reactions= pd.read_csv("Reactions .csv")
reactions
```

```python
reaction_type= pd.read_csv("ReactionTypes.csv")
reaction_type
```

### DATA CLEANING PROCESS

```python
#Dropping irrelevant columns not needed for the analysis
content.drop(["URL", "Unnamed: 0","User ID"], axis= 1, inplace= True)
reactions.drop(["User ID","Unnamed: 0"], axis= 1, inplace= True)
reaction_type.drop("Unnamed: 0", axis= 1, inplace= True)
```

```python
content
```

```python
reactions
```

```python
reaction_type
```

```python
## Renaming the columns names
content.rename(columns={"Type": "Content Type"}, inplace= True)
reactions.rename(columns={"Type": "Reaction Type"}, inplace= True)
reaction_type.rename(columns={"Type": "Reaction Type"}, inplace= True)
```

```python
content
```

```python
reactions
```

```python
reaction_type
```

```python
##converting reactions datetime to year
reactions["Datetime"]= pd.to_datetime(reactions["Datetime"])
reactions["year"]= reactions["Datetime"].dt.year
reactions['Month'] = reactions['Datetime'].dt.month
reactions['Month'] = reactions['Month'].apply(lambda x: calendar.month_name[x])
```

```python
reactions
```

```python
##Checking Null values in each dataset
reactions.isna().sum()
```

```python
reaction_type.isna().sum()
```

```python
content.isna().sum()
```

```python
#Dropping null values in reaction dataset
reactions.dropna(subset= "Reaction Type", axis= 0, inplace= True)
```

```python
content["Category"].value_counts()
```

```python
##Replacing the category of the content dataset
content['Category']= content['Category'].str.replace('"', '')
```

```python
category_mapping = {
    'Animals': 'animals',
    'Healthy Eating':'healthy eating',
    'Food':'food',
    'Veganism':'veganism',
    'Culture':'culture',
    'Technology':'technology',
    'Travel':'travel',
    'Education':'education',
    'Studying':'studying',
    'Soccer':'soccer',
    'Science':'science',
    'Fitness':'fitness',
    'Public Speaking':'public speaking'
}
content['Category'] = content['Category'].replace(category_mapping)
```

```python
content["Category"].value_counts()
```

### Merging the 3 Datasets

```python
x= reactions.merge(content, on= "Content ID")
x
```

```python
##Creating a new dataframe
content_df= x.merge(reaction_type, on='Reaction Type')
content_df
```

```python
content_df.to_csv('SocialbuzzCleaned_Data.csv')
```

## EXPLORATORY DATA ANALYSIS

## QUESTION 1
### What is Social Buzz Top 5 most popular content category 

```python
top_categories = content_df['Category'].value_counts().head(5).index
```

```python
plt.figure(figsize=[8, 5])
ax = sns.countplot(data=content_df, x='Category', order=top_categories)

plt.xlabel('Category')
plt.ylabel('Score')
plt.title('Social Buzz Top 5 content category Distribution')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=9, fontweight='light')

plt.show()
```

```python
top_categories = content_df['Sentiment'].value_counts()
plt.figure(figsize=[8, 5])
ax = sns.countplot(data=content_df, x='Sentiment')

plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.title('Social Buzz content Reaction Setiment Distribution')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=9, fontweight='light')
plt.show()
```

**Observation:** From the (Barchart and pie chart above),Social Buzz top 5 most popular content categories post are animals, science, healthy eating, food and technology.
Animals and science are the two most popular categories of content (with 21.5% and 20.4% respectively).

## QUESTION 2
### What is the Content Reaction Setiment for Social Buzz Contents?

```python
top_categories = content_df['Sentiment'].value_counts()
plt.figure(figsize=[8, 5])
ax = sns.countplot(data=content_df, x='Sentiment')

plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.title('Social Buzz content Reaction Setiment Distribution')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=9, fontweight='light')
plt.show()
```

```python
content_type_category = content_df['Sentiment'].value_counts().head(5)
labels = content_type_category.index
sizes = content_type_category.values

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Social Buzz content Reaction Setiment Distribution')

plt.show()
```

**Observation:** The social buzz content reaction setiment distribution shows that there is a larger positive reactions to their content posts and lower percentage of Negative and neutral reaction.

## QUESTION 3
### What is Social Buzz most popular content type 

```python
content_type_category = content_df['Content Type'].value_counts().head(5)
labels = content_type_category.index
sizes = content_type_category.values

plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Social Buzz Content Type')

plt.show()
```

**Observation:** From the pie chart above,Social Buzz most popular content types are photo,video,GIF and audio.
Photo are videos has the highest percentage (with 26.8% and 25.4% respectively).It is also seen that most people have a lower interest in audio content type.


## QUESTION 4
### What month has the most content post ?

```python
plt.figure(figsize=[11, 5])
ax = sns.countplot(data=content_df, x='Month', order=content_df["Month"].value_counts(ascending=True).index)

plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Month with the most Content Post')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=9, fontweight='light')

plt.show()
```

**Observation:** From the bar chart May, Jan, Aug, Dec are the months people post the most. However,January and May rank top 2 month where users post the most.This aligns with seasonal trends of social media users that feel the need to reconnect with people after calendar events such as Christmas.

## QUESTION 5
### What year has the most content post ?

```python
plt.figure(figsize=[8, 5])
ax = sns.countplot(data=content_df, x='year'),content_df['year'].value_counts()
plt.xlabel('year')
plt.ylabel('count')
plt.title('Social Buzz content views by Year');
```

**Observation:** We can see from the chart that there is a reduction in the content views from 2020 to 2021.
